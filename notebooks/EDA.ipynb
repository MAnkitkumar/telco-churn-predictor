# Telco Customer Churn — Exploratory Data Analysis
Understanding the data before modelling.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
df = pd.read_csv('../data/processed/churn_cleaned.csv')
print(df.shape)
df.head()
```

```python
# Churn rate
churn_rate = df['Churn Value'].mean()
print(f'Churn rate: {churn_rate:.1%}')

fig, ax = plt.subplots(figsize=(5, 4))
df['Churn Value'].value_counts().plot.pie(
    labels=['No Churn', 'Churn'],
    autopct='%1.1f%%',
    colors=['#3fb950', '#f85149'],
    ax=ax
)
ax.set_title('Churn Distribution')
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Churn by Contract Type
fig, ax = plt.subplots(figsize=(7, 4))
ct = df.groupby('Contract')['Churn Value'].mean().sort_values(ascending=False)
ct.plot.bar(color=['#f85149', '#e3b341', '#3fb950'], ax=ax)
ax.set_title('Churn Rate by Contract Type')
ax.set_ylabel('Churn Rate')
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
plt.tight_layout()
plt.show()
```

```python
# Tenure vs Churn
fig, ax = plt.subplots(figsize=(8, 4))
df[df['Churn Value']==1]['Tenure Months'].hist(bins=30, alpha=0.7, label='Churn', color='#f85149', ax=ax)
df[df['Churn Value']==0]['Tenure Months'].hist(bins=30, alpha=0.7, label='No Churn', color='#3fb950', ax=ax)
ax.set_title('Tenure Distribution by Churn')
ax.set_xlabel('Tenure (months)')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Monthly Charges vs Churn
fig, ax = plt.subplots(figsize=(7, 4))
sns.boxplot(data=df, x='Churn Value', y='Monthly Charges',
            palette={0: '#3fb950', 1: '#f85149'}, ax=ax)
ax.set_xticklabels(['No Churn', 'Churn'])
ax.set_title('Monthly Charges vs Churn')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
fig, ax = plt.subplots(figsize=(10, 7))
num_cols = df.select_dtypes(include='number').columns
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', ax=ax)
ax.set_title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```
